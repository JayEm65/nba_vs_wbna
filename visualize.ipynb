```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df_defense = pd.read_csv('./extracted_data/wnba_top_50_offensive_per.csv')
sns.set_theme()

#Graph creation
sns.barplot(x='Player', y='O-PER', data=df_defense)

# output settings
plt.xticks(rotation=90)  #Rotation on X
plt.title('O_PER per player') 
plt.tight_layout()  #output setting

plt.show()

```
